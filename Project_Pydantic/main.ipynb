```python
from pydantic import ValidationError, Field
```

```python
class Product(BaseModel):
    id: int
    price: float
    in_Stock: bool
```

```python
try:
    p = Product(id="A1", price = "wow", in_Stock ="yes")
except ValidationError as e:
    print(e)

# Think of it like:
    # ValidationError = the thrown ball
    # try-except = the person catching it
# The catcher prevents the crash.
```

```python
class SignUp(BaseModel):
    uname: str = Field(min_length=3,max_length=12)
    age : int = Field(ge=18, le = 60)
    Score : float = Field(default=0.0, ge=0, le=100)
```

```python
try:
    s = SignUp(uname="SN", age="2", Score="wow")
except ValidationError as e:
    print(e)
```

```python
class Address(BaseModel):
    city: str
    pincode: int
```

```python
class Customer(BaseModel):
    name: str
    address: Address
```

```python
Payload = {
    "name":"Srinithya",
    "address" : {"city" :"Chennai", "pincode":"600075"}
}
```

```python
c = Customer(**Payload)
print(c)
```

```python
print(c.address.pincode, type(c.address.pincode))
```
